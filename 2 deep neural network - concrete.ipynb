- By stacking layers inside a `Sequential` model, I can build networks. 
- By adding an activation function after the hidden layers, the network is able to learn more complex(non-linear) relationships in the data. 

Purpose of this coding exercise: build a neural network with several hidden layers and explore some activation functions beyond ReLU. 

```python
import pandas as pd 

concrete = pd.read_csv('concrete.csv')
concrete.head()
```

## 1. Input shape
Target: the column `'CompressiveStrength'`. 

Inputs: the remaining columns

```python
concrete.shape
```

```python
input_shape = [8]
```

## 2. Define a Model with Hidden Layers
Now create a model with three hidden layers, each having 512 units and the ReLU activation. 
- output layer: one unit and no activation
- input_shape: an argument to the first layer

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    # the hidden ReLU layers
    layers.Dense(units=512, activation='relu', input_shape=[8]),
    layers.Dense(units=512, activation='relu'),
    layers.Dense(units=512, activation='relu'),
    # the linear output layer
    layers.Dense(units=1),
])
```

## 3. Activation Layers
Also, we can define the activation in its own `Activation` layer as following code.

```python
model = keras.Sequential([
    layers.Dense(units=32, input_shape=input_shape),
    layers.Activation('relu'),
    layers.Dense(units=32),
    layers.Activation('relu'),
    layers.Dense(units=1)
])
```

## 4. Alternatives to ReLU
- Changing activation names: `'relu'`, `'elu'`, `'selu'`,`'swish'`
- run the cell to see the graph
- Sometimes, one activation performs better than another
- ReLU activation tends to do well on most problems, so it is a good start with.

### ReLU

```python
import tensorflow as tf 
import matplotlib.pyplot as plt

# Activation: ReLU
activation_layer = layers.Activation('relu')

x = tf.linspace(-3.0, 3.0, 100)
y = activation_layer(x) 

plt.figure(dpi=100)
plt.plot(x,y)
plt.xlim(-3,3)
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
